```python
import pandas as pd
from google_play_scraper import reviews_all, Sort
import numpy as np
```

```python
us_review = reviews_all(
    'com.HoYoverse.hkrpgoversea',
    sleep_milliseconds=0,
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT)
```

```python
df_trv = pd.DataFrame(np.array(us_review), columns=['review'])
df_trv = df_trv.join(pd.DataFrame(df_trv.pop('review').tolist()))  # tolist:mengubah series menjadi list
```
